```python
import pandas as pd
import numpy as np
import string
```

```python
df = pd.DataFrame({"word": pd.read_fwf('raw_words.txt')["AA"].tolist()})
df.dropna(subset=['word'], inplace=True)
df["word"] = df["word"].astype(str)
```

```python
score_df = pd.DataFrame(pd.read_excel('scrabble_scoring.xlsx'))
```

```python
#creating alt columns for quicker lookup
df["length"] = df["word"].str.len()
```

```python
#render time ~ 15-20 mins

# counting each letter per word
letter_counts = []
scores = []
for i in range(len(df)):
    word = df["word"].iloc[i]
#     letters = {char : 0 for char in string.ascii_uppercase}
    letters = [0] * 26
    score = 0
    for char in word:
        score += score_df[score_df["Letter"] == char]["Score"].iloc[0]
        letters[ord(char.upper()) - ord("A")] += 1
    letter_counts.append(letters)
    scores.append(score)
    

# adding the letter counts as columns within dataframe
formatted_counts = list(zip(*letter_counts))
for char in string.ascii_uppercase:
    df[char] = list(formatted_counts[ord(char) - ord("A")])

df["Score"] = scores

df
```

```python
df.to_excel('data.xlsx', sheet_name="words")  
```
