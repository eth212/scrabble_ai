```python
def print_board(board):
    for i in range(15):
        print(board[i])
        
def place_board(board, word, position, direction):
    r, c = position[0], position[1]
    if direction == 'r' and (c + len(word)) <= 15:
        #checking if valid 
        for i in range(len(word)):
            if board[r][c + i] is None:
                board[r][c + i] = word[i]
            else:
                print("invalid")
                break
        
    elif direction == 'd' and (r + len(word)) <= 15:
        #checking if valid
        print(len(word),r,c)
        for i in range(len(word)):
            if board[r + i][c] is None:
                board[r + i][c] = word[i]
            else:
                print("invalid")
                break
    else:
        print("invalid")
    return board
```
